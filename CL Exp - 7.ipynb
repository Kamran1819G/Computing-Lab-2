```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load a dataset
tips = sns.load_dataset("tips")

# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Show the plot
plt.show()
```

```python
# Load a dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="region", data=fmri)

# Show the plot
plt.show()
```

```python
# Load a dataset
titanic = sns.load_dataset("titanic")

# Create a bar plot
sns.barplot(x="age", y="fare", data=titanic)

# Show the plot
plt.show()
```

```python
# Load a dataset
iris = sns.load_dataset("iris")

# Create a histogram
sns.histplot(data=iris, x="sepal_length")

# Show the plot
plt.show()
```

```python
# Load a dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create the required format for the heatmap
sns.heatmap(flights.pivot_table(index="month", columns="year", values="passengers"), cmap="Reds", cbar_kws={"ticks":np.arange(0, 1000, 200)})
plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Number of Flights per Month")
plt.show()
```
